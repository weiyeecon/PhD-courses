```python
import matplotlib.pyplot as plt

import networkx as nx
```

```python
class GraphVisualization:
    def __init__(self):
        self.visual = []
        
    def addEdge(self, a, b):
        temp = [a, b]
        self.visual.append(temp)
        
    def visualize(self):
        G = nx.Graph()
        G.add_edges_from(self.visual)
        nx.draw_networkx(G)
        plt.show()
```

```python
G = GraphVisualization()

G.addEdge('$v_3$','$v_5$')
G.addEdge('$v_5$','$v_1$')
G.addEdge('$v_5$','$v_2$')
G.addEdge('$v_1$','$v_6$')
G.addEdge('$v_2$','$v_6$')
G.addEdge('$v_6$','$v_4$')

G.visualize()
plt.savefig('(b)1.png')
```
